```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset of Penguins
df = sns.load_dataset("penguins")
df
```

```python
# Bar Plots

sns.barplot(data=df, x="species", y="body_mass_g", hue="sex")
plt.title("Bar Plots : Body Mass by Species and Sex", color = 'r')
plt.show()
```

```python
sns.barplot(data=df, x= 'island',y= 'bill_length_mm', hue = 'sex',width=0.2)
plt.title("Bar Plots : ",color='g')
plt.grid()
plt.show()
```

```python
# Order Parameter

v = ["Dream", "Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data=df, order = v, width=0.3, hue = 'sex', palette='prism', saturation=0.75)
# sns.suptitle("Bar Plots", color='r', y=1.01)
plt.title("Bar Plots", color='r', y=1.01)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Penguins dataset
df = sns.load_dataset("penguins")

# Order of the islands
v = ["Dream", "Torgersen", "Biscoe"]

# Create the bar plot with 'hue' assigned
sns.barplot(x="island",y="bill_length_mm",data=df,order=v,width=0.3,hue="sex",hue_order=["female", "male"],palette="prism")
# hue_order=["female", "male"]  # For this not showing plots
# Add title
plt.title("Bar Plots", color="r", y=1.01)
# sns.suptitle("Bar Plots", color='r', y=1.01)

# Show the plot
plt.show()
```
