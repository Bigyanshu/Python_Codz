```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset of Penguins
df = sns.load_dataset("penguins")
df

# create variables
v1 = [2, 4, 6]
v2 = [1, 3, 5]
sns.lineplot(x=v1, y=v2,markers='o')
plt.title("Line Plot", color= 'r')
plt.legend(title="Line Plot")
plt.show()
```

```python
# Load the Penguins dataset
df = sns.load_dataset("penguins")

# Preview the dataset
print(df.head(10))

# Create a line plot
sns.lineplot(
    x="bill_length_mm", 
    y="bill_depth_mm", 
    data=df, 
    palette="rocket_r", 
    hue="sex",
    markers="o"
)

# Add grid
plt.grid()
# Show the plot
plt.show()
```

```python
# Line Plot

sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = df, hue = 'sex', palette="rocket_r", style= 'sex', markers= ["o",">"])
plt.title("Line Plot", color= 'b')
plt.show()
```

```python
# Sample Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 2.5, 30]

sns.lineplot(x=x, y=y,color='r')
plt.title("Line Plots", color='b')
plt.grid()
plt.xlabel("X-Axis--->")
plt.ylabel("Y-Axis--->")
plt.show()
```
