```python
# Pair Plots

# Box Plot

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset of Penguins
df = sns.load_dataset("penguins")
df
```

```python
# Pair Plots

sns.pairplot(df, hue = "species")
plt.suptitle("Pair Plots : Pair_Wise Relations", color= 'b',y = 1.03)
plt.grid()
plt.show()
```
