```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 4, 5]

plt.plot(x,y,color = 'r')
plt.xlabel ('X-axis')
plt.ylabel("Y-axis")
plt.title("Linear Graph")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [9,8,7,6,1]
c = 'g'
plt.title("Linear Graph")
plt.plot(x,y,c)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# plt.show()
```

```python
line = plt.plot([1, 2, 3], [4, 5, 6])  # Plot a line
print(line)
```

```python
lines = plt.plot([1,9,3,7],[3,7,4,8])
plt.plot(x, marker = 'o')
plt.plot(y, marker = '>')
print(lines)
```
