```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for the pie chart

x = [10, 30, 50, 10, 10]  # Numerical values
y = ['Sanskrit', 'MIL', 'MATH', 'Computer', 'Chemistry']  # Labels

plt.title('Pie Chart',color= 'b')
# Create the pie chart
plt.pie(x, labels=y, autopct='%1.1f%%')  # Add percentages with autopct
plt.legend(title="Subjects")
# Display the chart
plt.show()
```
