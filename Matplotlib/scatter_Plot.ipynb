```python
import matplotlib.pyplot as plt
```

```python
x = [1,3,5,2,4]
y = [4,6,7,9,0]

c ='g'
plt.title("Scatter Plot",color='r')
plt.scatter(x,y,color=c)
plt.xlabel('Months --->',color='b')
plt.ylabel('Years --->',color='b')
plt.legend(x,y )
# plt.axis('off')
plt.show()
plt.grid()
```
